# Caos e mappa logistica

La  logistic map è il modello più classico  per introdurre le proprietà di un sistema dinamico caotico - ovvero nel quale lo stato di una variabile ha un potere predittivo quasi nullo sullo stato della stessa variabile nei periodi successivi. I sistemi caotici si differenziano da sistemi che convergono a un punto fisso o da cicli che si ripetono con periodicità data. Un altro modo di definire un sistema caotico è che ha un ciclo di periodo infinito.


Modelli caotici sono stati utilizzati per spiegare numerosi fenomeni. Ad esempio, potrebbero spiegare i fenomenno delle fibrillazioni cardiache. La perdita della regolarità periodica del battito cardiaco potrebbe esser dovuta a un cambiamento in un parametro che ne regola la dinamica. Perché non l'ora cui ci si addormenta, che dipende dai ritmi circadiani? 

La logistic map è molto semplice: x(t+1)=r * x(t) *(1-x(t)) 
Essenzialmente raooresenta un fenomeno di crescita esponenziale (x al tempo t+1 è paria  a r volte x al tempo t, con r>1) rallentata da un processo di saturazione (più x si avvicina a 1, più riduce la propria corsa).
Ad esempio, puo' esser usata per rappresentare la crescita esponenziale di una popolazione limitata malthusanamente da risorse finite. Oppure puo' essere usata per rappresentare la adozione di un nuovo prodotto.

Il tipico comportamento della mappa logistica è quindi di generare una curva sigmoide, a "S". Tutavia, al crescere del valore del parametro *r* che ne regola la crescita, la mappa logistica assume comportamenti più complessi. Inzialmente  produce fluttuazioni cicliche di complessità sempre crescente. Oltre un valore critico di 3.57 cambia ulteriormente comportamento e mostra una dinamica caotica. 

La logisitc map è anche un esempio semplice per introdurre la programmazione di un sistema dinamico. Infatti ha una sola variabile ed è una equazione alle differenze finite - la cui dinamica cioé si svolge per passi discreti, come nella computazione fatta da un computer. 

Innanzitutto devi importare una libreria che servirà per la grafica - dalle un nickname (plt) per semplificarne l'utilizzo

```
import matplotlib.pyplot as plt
import numpy as np
```

 La funzione è la classica logistic map: x=r*x*(1-x)


```
def l(x):
  x=r*x*(1-x)
  return x
```

**Inizializzazione**. Devi fissare il parametro della equazione, r - con valori compresi fra 0.01 e 4. 
x è invece il valore iniziale della variabile.

Come puoi vedere, la funzione genera un andamento a "U" rovesciata nell'intervallo di valori di x fra 0 e 1. Prova con divdersi valori di r!

```
r=1.8
x = np.linspace(0,1)
y=l(x)
plt.plot(x,y,'k')
plt.show()
```

Potete immaginare fenomeni sociali che si comportino così?

**Ora fai correre la funzione**. Prima crea un vettore vuoto (values) dove metterai via via i valori successivi della funzione.
Poi crea un loop di durata 200 periodi che aggiorni per 200 volrte successive il valore di x usandio al funzione l(x).
poi "appendi" al vetore values il nuovo valore...
Infine crea una figlura di dimesione adegiata..
e plotta i valori accumulati nel vettore values. 
La cosa interessante è provare diversi valori di r. 
Suggerisco di cominciare da r=0.9. Dovresti vedere la funzione declinare  
Prova r=1.1, ora emerge la classica logistica della diffusione dei prodotti.
Con r=2.9.le coswe si fanno interessanti.Dopo un rapida crescita la funzione entra in ciclo con piccole fluttuazioni che si attenuano rapidamente.
Con r=3.1 il ciclo si stabilizza con periodo 2.
Con r=3.5 il ciclo si complica (periodo 4). Con r=3.7 è parecchio complesso. Sei entrata in un regime caotico!

```

r=2.9
x=0.001
values=[]
for i in range(200):
  x=l(x)
  values.append(x)
plt.figure(figsize=(30,10))
plt.plot(values)

```

# L'effetto farfalla.
L'imprevedibilità delle dinamiche caotiche fa sì che piccoli spostamenti dello stato iniziale producano dinamiche del tutto diverse. È il cosiddetto "butterfly effect": un battito d'ala di una farfalla in Cina puo'produrre un tornado nel Texas! Il codice che segue sovrappone due curve con uno stesso valore di r (3.99) che induce dinamiche caotiche e con una leggerissima differenza (0.000001) nello stato inziale (il battito d'ali). Come puoi vedere,dopo poco più di una decina di passi le due traiettore divergono e divengono rapidamente incorrelate fra loro. 

```
r=3.99
x1=0.200001
x2=0.200002
values1=[]
values2=[]
for i in range(50):
  x1=l(x1)
  values1.append(x1)
for i in range(50):
  x2=l(x2)
  values2.append(x2)

plt.figure(figsize=(30,10))
plt.plot(values1)
plt.plot(values2)
plt.show()
```
